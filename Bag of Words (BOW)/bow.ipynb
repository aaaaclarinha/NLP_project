#### Plain of Frequency

```python
import spacy
import pandas as pd
from scipy import spatial
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
data = pd.read_excel('C:/JupyterLab/github_repos/NLP_project/Dados/df_completo.xlsx')
```

```python
corpus = list(data['Abstract'][:2])
print(corpus)
```

```python
vectorizer = CountVectorizer()
```

```python
bow = vectorizer.fit_transform(corpus)
```

```python
# View features (tokens).
print(vectorizer.get_feature_names_out())

# View vocabulary dictionary.
vectorizer.vocabulary_
```

```python
print(bow)
```

#### BOW BinÃ¡rio

```python
spacy_tokenizer = lambda doc : [t.text for t in nlp(doc) if not t.is_punct]
```

```python
vectorizer = CountVectorizer(tokenizer=spacy_tokenizer, lowercase=False, binary=True)
bow = vectorizer.fit_transform(corpus)
```

```python
print(vectorizer.get_feature_names_out())
vectorizer.vocabulary_
```

```python
print('A dense representation like we saw in the slides.')
print(bow.toarray())
print()
print('Indexing and slicing.')
print(bow[0])
print()
print(bow[0:2])
```
