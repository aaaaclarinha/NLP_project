```python
import pandas as pd
import re
import spacy

# Assuming 'nlp' is initialized elsewhere in your code
# nlp = spacy.load("en_core_web_sm")

# List of common cancer and tumor types
cancer_types_list = ["breast", "lung", "prostate", "colon", "rectal", "skin", "leukemia", "lymphoma", "melanoma", "ovarian", "pancreatic", "brain", "bone", "liver", "kidney", "cervical", "uterine", "bladder", "testicular", "thyroid"]

# Compile regex pattern for cancer and tumor types
cancer_types_pattern = re.compile(r"\b(?:{})(?![A-Z])".format('|'.join(cancer_types_list)), re.IGNORECASE)

dados_extraidos = pd.DataFrame(columns=['Cancer Type', 'Abstract'])

i = -1
for abstract in dados['Abstract']:
    i += 1
    doc = nlp(str(abstract))
    for sentence in doc.sents:
        cancer_types = re.findall(cancer_types_pattern, str(sentence))
        if len(cancer_types) != 0:
            for cancer_type in cancer_types:
                if cancer_type[0].islower():  # Check if the first character is lowercase
                    print('index', i, 'tipo de cancer:', cancer_type) # Print for verification
                    tmp_df = pd.DataFrame({'Cancer Type': [cancer_type.lower()],  # Convert to lowercase for consistency
                                           'Abstract': [str(sentence)]})
                    dados_extraidos = pd.concat([dados_extraidos, tmp_df], ignore_index=True)

# Print or use dados_extraidos DataFrame as needed
print(dados_extraidos)
```
