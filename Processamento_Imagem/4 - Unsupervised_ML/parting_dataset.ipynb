```python
import os
import shutil

def copy_first_images(quantity_images,src_root, dest_root):
    if not os.path.exists(dest_root):
        os.makedirs(dest_root)

    for category in ['training_set', 'test_set']:
        src_category_path = os.path.join(src_root, category)
        dest_category_path = os.path.join(dest_root, category)
        
        if not os.path.exists(dest_category_path):
            os.makedirs(dest_category_path)

        for subfolder in ['lung_aca', 'lung_n', 'lung_scc']:
            src_subfolder_path = os.path.join(src_category_path, subfolder)
            dest_subfolder_path = os.path.join(dest_category_path, subfolder)
            
            if not os.path.exists(src_subfolder_path):
                print(f"Source subfolder does not exist: {src_subfolder_path}")
                continue

            if not os.path.exists(dest_subfolder_path):
                os.makedirs(dest_subfolder_path)
                
            images = sorted([img for img in os.listdir(src_subfolder_path) if img.lower().endswith(('.png', '.jpg', '.jpeg'))])
            for image in images[:quantity_images]:
                src_image_path = os.path.join(src_subfolder_path, image)
                dest_image_path = os.path.join(dest_subfolder_path, image)
                shutil.copyfile(src_image_path, dest_image_path)
                print(f"Copied {src_image_path} to {dest_image_path}")

# Define source and destination root directories
src_root = '../Images_data/'
dest_root = '../Images_data/test_set_reduced'
quantity_images = 100

copy_first_images(quantity_images,src_root, dest_root)
```
